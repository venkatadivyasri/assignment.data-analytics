```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_path = ("C:\Users\Chinni Krishna\Downloads\data\sample_sales_data (1).csv" ) # Ensure this file exists in the same directory
df = pd.read_csv(data_path)

# Display the first few rows of the dataset
print("Dataset Preview:")
print(df.head())

# Add a Revenue column
df['Revenue'] = df['Quantity'] * df['Price'] - df['Discount']

# ---- Revenue Analysis ----
# Total revenue
total_revenue = df['Revenue'].sum()
print(f"\nTotal Revenue: {total_revenue}")

# Revenue by product
product_revenue = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
print("\nRevenue by Product:")
print(product_revenue)

# ---- Visualizing Revenue by Product ----
plt.figure(figsize=(8, 5))
plt.bar(product_revenue.index, product_revenue.values, color='skyblue')
plt.title('Revenue by Product', fontsize=14)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.show()

# ---- Customer Insights ----
# Total spend by customer
customer_spend = df.groupby('Customer_ID')['Revenue'].sum().sort_values(ascending=False)
print("\nTotal Spend by Customer:")
print(customer_spend)

# ---- Sales Trends ----
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')
monthly_revenue = df.groupby('Month')['Revenue'].sum()

# Visualize monthly revenue
plt.figure(figsize=(10, 6))
plt.plot(monthly_revenue.index.astype(str), monthly_revenue.values, marker='o', color='green')
plt.title('Monthly Revenue Trend', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# ---- Discount vs Revenue Correlation ----
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Discount', y='Revenue', data=df, color='orange')
plt.title('Discount vs Revenue', fontsize=14)
plt.xlabel('Discount', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.show()

# Summary
print("\nAnalysis Complete!")
```
